### 1.3. <a id='toc1_3_'></a>–°–µ—Ä–≤–∏—Å–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏                 [&#8593;](#toc0_)

```python
def print_table_about_nans(data: pd.DataFrame) -> None:
    table = PrettyTable()
    table.field_names = ['Feature', 'Missing values count']
    
    missing_info = data.isna().sum().sort_values()
    cols = missing_info.index.to_list()
    for col in cols:
        count = missing_info[col]
        color = 'green' if count == 0 else 'red'
        s = f'{count} ({count / len(data):.2%})'
        table.add_row([col, colored(s, color)])
    
    print(table)
```

```python
def draw_plots_about_column(df: pd.DataFrame,
                            column: str,
                            column_description: str,
                            bins: int = 20
                            ) -> None:
    _, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))
    
    sns.violinplot(df, x=column, ax=ax1, orient='h')
    ax1.set_xlabel(column_description)
    ax1.grid()
    
    sns.boxplot(df, x=column, ax=ax2)
    ax2.set_xlabel(column_description)
    ax2.grid()
    
    sns.histplot(df, x=column, bins=bins, ax=ax3)
    ax3.set_xlabel(column_description)
    ax3.grid()
    ax3.set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
    
    ax4.set_visible(False)
    
    plt.suptitle(f'–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –∫–æ–ª–æ–Ω–∫–µ `{column_description}`', fontsize=18)
    plt.tight_layout()

    plt.show()
```

```python
def plot_pie_chart(series: pd.Series) -> None:
    _table = series.value_counts()
    data, labels = _table.values, _table.index
    plt.pie(data, labels=labels, autopct='%.0f%%')
    plt.title(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤ —Å—Ç–æ–ª–±—Ü–µ `{series.name}`')
    plt.show()
```

```python
def get_dataframe(paths: list[str]) -> pd.DataFrame:
    for _path in paths:
        if not exists(_path) and not _path.startswith('http'):
            continue
        
        try:
            df = pd.read_csv(_path)
        except:
            continue
        
        if df is None:
            continue
        
        return df
    
    raise FileNotFoundError('No paths are valid and/or for correct csv file.')
```

```python
def smape_function(y_true: np.ndarray, y_pred: np.ndarray) -> float:    
    return np.mean(np.abs(y_true - y_pred) / ((np.abs(y_true) + np.abs(y_pred)) / 2)) * 100

smape = make_scorer(smape_function, greater_is_better=False)
```

```python
def get_plots_about_residuals(y_true: np.array, y_pred: np.array) -> None:
    res = y_true - y_pred
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 5))
    
    sns.histplot(res, ax=ax1)
    ax1.set_xlabel('–û—Å—Ç–∞—Ç–∫–∏')
    ax1.set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
    ax1.axvline(x=0, color='green', lw=3, ls='-', label='–ù—É–ª–µ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ')
    ax1.axvline(x=res.mean(), color='red', lw=3, ls='--', label='–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤')
    ax1.legend()
    
    sns.scatterplot(x=y_pred, y=res, ax=ax2)
    ax2.set_xlabel('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏–µ')
    ax2.set_ylabel('–û—Å—Ç–∞—Ç–∫–∏')
    
    fig.suptitle('–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ–± –æ—Å—Ç–∞—Ç–∫–∞—Ö')
```

```python
def draw_pair_pie_plot(
    first_df: pd.DataFrame,
    first_df_name: str,
    second_df: pd.DataFrame,
    second_df_name: str,
    col: str
    ) -> None:
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
    
    table1 = first_df[col].value_counts()
    data, labels = table1.values, table1.index
    ax1.pie(data, labels=labels, autopct='%.0f%%')
    ax1.set_title(f'–î–∞—Ç–∞—Ñ—Ä–µ–π–º `{first_df_name}`')
    
    table2 = second_df[col].value_counts()
    data, labels = table2.values, table2.index
    ax2.pie(data, labels=labels, autopct='%.0f%%')
    ax2.set_title(f'–î–∞—Ç–∞—Ñ—Ä–µ–π–º `{second_df_name}`')
    
    fig.suptitle(f'–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤ —Å—Ç–æ–ª–±—Ü–µ `{col}`')
    plt.show()
    
```

### 6.2. <a id='toc6_2_'></a>–ü–∞–π–ø–ª–∞–π–Ω—ã                 [&#8593;](#toc0_)

```python
ohe_pipeline = Pipeline(
    [
        (
            'SimpleImputer_before_ohe_1',
            SimpleImputer(strategy='most_frequent', missing_values=np.nan)
        ),
        (
            'SimpleImputer_before_ohe_2',
            SimpleImputer(strategy='most_frequent', missing_values=' ')
        ),
        (
            'ohe',
            OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
        )
    ]
)
```

```python
ord_pipeline = Pipeline(
    [
        (
            'SimpleImputer_before_ord_1',
            SimpleImputer(strategy='most_frequent', missing_values=np.nan)
        ),
        (
            'SimpleImputer_before_ord_2',
            SimpleImputer(strategy='most_frequent', missing_values=' ')
        ),
        (
            'ord',
            OrdinalEncoder(
                categories=[
                    ['junior', 'middle', 'sinior'],
                    ['low', 'medium', 'high'],
                    ['no', 'yes'],
                    ['no', 'yes']
                ],
                unknown_value=np.nan,
                handle_unknown='use_encoded_value'
            )
        ),
        (
            'SimpleImputer_after_ord',
            SimpleImputer(strategy='most_frequent', missing_values=np.nan)
        ),
    ]
)
```

```python
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipeline, ohe_cols),
        ('ord', ord_pipeline, ord_cols),
        ('num', MinMaxScaler(), num_cols)
    ],
    remainder='passthrough'
)
```

```python
final_pipeline = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', DecisionTreeRegressor(random_state=RANDOM_STATE))
    ]
)
```

```python
final_pipeline
```

---

```python
param_grid = [
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(2,  5),
        'models__max_features': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    {
        'models': [KNeighborsRegressor(n_jobs=-1)],
        'models__n_neighbors': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    {
        'models': [LinearRegression(n_jobs=-1)],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    {
        'models': [SVR()],
        'models__C': [0.01, 0.1, 1, 10, 100],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    {
        'models': [RandomForestRegressor(random_state=RANDOM_STATE, n_jobs=-1)],
        'models__n_estimators': [100, 500, 1000],
        'models__min_samples_split': [2, 4],
        'models__min_samples_leaf': [1, 3],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]
```

---

```python
grid = GridSearchCV(
    final_pipeline,
    param_grid=param_grid,
    scoring=smape,
)
```
